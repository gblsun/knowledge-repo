```python
# Boxplot dos tempos
import matplotlib.pyplot as plt

dados = [120, 150, 180, 200, 250]
dados_ordenados = sorted(dados)

print("Dados ordenados:", dados_ordenados)
```

```python
# Quartis pelo método manual
minimo = 120
q1 = 135
q2 = 180
q3 = 225
maximo = 250

aiq = q3 - q1
limite_inferior = q1 - 1.5 * aiq
limite_superior = q3 + 1.5 * aiq

outliers = [x for x in dados_ordenados if x < limite_inferior or x > limite_superior]

print("Mínimo =", minimo)
print("Q1 =", q1)
print("Q2 =", q2)
print("Q3 =", q3)
print("Máximo =", maximo)
print("AIQ =", aiq)
print("Limite inferior =", limite_inferior)
print("Limite superior =", limite_superior)
print("Outliers =", outliers)
```

```python
# Boxplot com anotações
plt.figure(figsize=(11, 4))
plt.boxplot(dados_ordenados, vert=False, widths=0.5)

plt.title("Boxplot dos tempos")
plt.xlabel("Tempo (ms)")
plt.yticks([1], ["Tempos"])

plt.axvline(minimo, color='gray', linestyle='--', linewidth=1)
plt.axvline(q1, color='blue', linestyle='--', linewidth=1)
plt.axvline(q2, color='red', linestyle='--', linewidth=1)
plt.axvline(q3, color='blue', linestyle='--', linewidth=1)
plt.axvline(maximo, color='gray', linestyle='--', linewidth=1)

plt.text(minimo, 1.18, "Min\n120", ha='center')
plt.text(q1, 0.78, "Q1\n135", ha='center', color='blue')
plt.text(q2, 1.18, "Q2\n180", ha='center', color='red')
plt.text(q3, 0.78, "Q3\n225", ha='center', color='blue')
plt.text(maximo, 1.18, "Max\n250", ha='center')

plt.text((q1 + q3)/2, 1.33, "Caixa: 135 até 225", ha='center', color='green')

plt.xlim(100, 260)
plt.grid(axis='x', linestyle=':', alpha=0.5)

plt.show()
'''
    Essa figura está completamente errada
'''
```

```python
# Boxplot manual dos tempos
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle

minimo = 120
q1 = 135
q2 = 180
q3 = 225
maximo = 250
```

```python
# Desenho manual do boxplot
fig, ax = plt.subplots(figsize=(11, 4))

y = 1
altura = 0.5

# Caixa
caixa = Rectangle((q1, y - altura/2), q3 - q1, altura, fill=False, edgecolor='black')
ax.add_patch(caixa)

# Mediana
ax.vlines(q2, y - altura/2, y + altura/2, colors='red')

# Bigodes
ax.hlines(y, minimo, q1, colors='black')
ax.hlines(y, q3, maximo, colors='black')

# Extremidades dos bigodes
ax.vlines(minimo, y - 0.1, y + 0.1, colors='black')
ax.vlines(maximo, y - 0.1, y + 0.1, colors='black')

# Linhas auxiliares
ax.axvline(minimo, color='gray', linestyle='--', linewidth=1)
ax.axvline(q1, color='blue', linestyle='--', linewidth=1)
ax.axvline(q2, color='red', linestyle='--', linewidth=1)
ax.axvline(q3, color='blue', linestyle='--', linewidth=1)
ax.axvline(maximo, color='gray', linestyle='--', linewidth=1)

# Textos
ax.text(minimo, 1.18, "Min\n120", ha='center')
ax.text(q1, 0.78, "Q1\n135", ha='center', color='blue')
ax.text(q2, 1.18, "Q2\n180", ha='center', color='red')
ax.text(q3, 0.78, "Q3\n225", ha='center', color='blue')
ax.text(maximo, 1.18, "Max\n250", ha='center')
ax.text((q1 + q3)/2, 1.33, "Caixa: 135 até 225", ha='center', color='green')

ax.set_title("Boxplot manual dos tempos")
ax.set_xlabel("Tempo (ms)")
ax.set_yticks([1])
ax.set_yticklabels(["Tempos"])
ax.set_xlim(100, 260)
ax.set_ylim(0.5, 1.5)
ax.grid(axis='x', linestyle=':', alpha=0.5)

plt.show()
```
