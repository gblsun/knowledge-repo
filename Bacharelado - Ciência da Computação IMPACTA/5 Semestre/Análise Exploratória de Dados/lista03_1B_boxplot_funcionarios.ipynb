```python
# Boxplot dos salários

import numpy as np
import matplotlib.pyplot as plt

dados = [2500, 2800, 3000, 3200, 3500]
dados_ordenados = sorted(dados)

print("Dados ordenados:", dados_ordenados)
```

```python
# Cálculo dos quartis pelo método manual (mediana das metades)

q2 = dados_ordenados[len(dados_ordenados)//2]

parte_inferior = dados_ordenados[:len(dados_ordenados)//2]
parte_superior = dados_ordenados[len(dados_ordenados)//2 + 1:]

q1 = sum(parte_inferior) / len(parte_inferior)
q3 = sum(parte_superior) / len(parte_superior)

print("Q1 =", q1)
print("Q2 (mediana) =", q2)
print("Q3 =", q3)
```

```python
# Amplitude interquartil e limites para outliers

aiq = q3 - q1
limite_inferior = q1 - 1.5 * aiq
limite_superior = q3 + 1.5 * aiq

print("AIQ =", aiq)
print("Limite inferior =", limite_inferior)
print("Limite superior =", limite_superior)
```

```python
# Identificação de outliers

outliers = [x for x in dados_ordenados if x < limite_inferior or x > limite_superior]

print("Outliers =", outliers)
```

```python
# Resumo dos cinco números

minimo = min(dados_ordenados)
maximo = max(dados_ordenados)

print("Mínimo =", minimo)
print("Q1 =", q1)
print("Q2 =", q2)
print("Q3 =", q3)
print("Máximo =", maximo)
```

```python
# Boxplot

plt.boxplot(dados_ordenados, vert=False)
plt.title("Boxplot dos salários")
plt.xlabel("Salário (R$)")
plt.show()
```

```python
dados = [2500, 2800, 3000, 3200, 3500]

# Valores calculados manualmente
minimo = 2500
q1 = 2650
q2 = 3000
q3 = 3350
maximo = 3500
```

```python
# Boxplot com marcações verticais e rótulos

plt.figure(figsize=(11, 4))
plt.boxplot(dados, vert=False, widths=0.5)

plt.title("Boxplot dos salários")
plt.xlabel("Salário (R$)")
plt.yticks([1], ["Salários"])

# Linhas verticais nos pontos importantes
plt.axvline(minimo, color='gray', linestyle='--', linewidth=1)
plt.axvline(q1, color='blue', linestyle='--', linewidth=1)
plt.axvline(q2, color='red', linestyle='--', linewidth=1)
plt.axvline(q3, color='blue', linestyle='--', linewidth=1)
plt.axvline(maximo, color='gray', linestyle='--', linewidth=1)

# Rótulos
plt.text(minimo, 1.18, "Min\n2500", ha='center', color='black')
plt.text(q1, 0.78, "Q1\n2650", ha='center', color='blue')
plt.text(q2, 1.18, "Q2\n3000", ha='center', color='red')
plt.text(q3, 0.78, "Q3\n3350", ha='center', color='blue')
plt.text(maximo, 1.18, "Max\n3500", ha='center', color='black')

# Texto indicando a caixa
plt.text((q1 + q3)/2, 1.33, "Caixa: 2650 até 3350", ha='center', color='green')

plt.xlim(2400, 3600)
plt.grid(axis='x', linestyle=':', alpha=0.5)

plt.show()
```

```python
# Quartis pelo método padrão do Python
q1 = np.percentile(dados, 25)
q2 = np.percentile(dados, 50)
q3 = np.percentile(dados, 75)

aiq = q3 - q1
limite_inferior = q1 - 1.5 * aiq
limite_superior = q3 + 1.5 * aiq

minimo = min(dados)
maximo = max(dados)

print("Mínimo =", minimo)
print("Q1 =", q1)
print("Q2 =", q2)
print("Q3 =", q3)
print("Máximo =", maximo)
print("Caixa:", q1, "até", q3)
```
